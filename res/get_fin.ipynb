```python
# pip install yfinance
import yfinance as yf
import pandas as pd
df= pd.read_csv("./banks_list_csv.csv")
ticker = df["Ticker"]
total_stock_price_m = pd.DataFrame()
for i in ticker:
    aapl = yf.Ticker(i)
    hist = aapl.history(start="2018-01-31",end="2024-07-05",interval="1mo")
    data=hist.resample('M').last()["Close"]
    total_stock_price_m[i]=data

total_stock_price_m
total_stock_price_m.to_csv("./total_stock_price_m.csv")
```
